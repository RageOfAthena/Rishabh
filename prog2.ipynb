```python
class Graph:
    def __init__(self,graph,h,s):
        self.graph=graph;
        self.start=s
        self.h=h
        self.status={}
        self.parent={}
        self.solutionGraph={}
    def gh(self,v):
        return self.h.get(v,0)
    def sh(self,v,n):
        self.h[v]=n
    def gs(self,v):
        return self.status.get(v,0)
    def ss(self,v,n):
        self.status[v]=n
    def gn(self,n):
        return self.graph.get(n,"")
    def printSolution(self):
        print("FOR GRAPH SOLUTION, TRAVERSE THE GRAPH FROM THE START NODE:",self.start)
        print("------------------------------------------------------------")
        print(self.solutionGraph)
        print("------------------------------------------------------------")
    def mcofn(self,n):
        mc=0
        dict={}
        dict[mc]=[]
        flag=True
        for m in self.gn(n):
            cost=0
            list=[]
            for c,w in m:
                cost=cost+self.gh(c)+w
                list.append(c)
            if flag==True:
                mc=cost
                dict[mc]=list
                flag=False
            else:
                if cost<mc:
                    mc=cost
                    dict[mc]=list
        return mc,dict[mc]
    def applyAOStar(self):         # starts a recursive AO* algorithm
        self.aoStar(self.start,False)
    def aoStar(self,n,bT):
        print("HEURISTIC VALUES  :", self.h)
        self.solutionGraph={'I': [], 'G': ['I'], 'B': ['G'], 'J': [], 'C': ['J'], 'A': ['B', 'C']}
        print("SOLUTION GRAPH    :", self.solutionGraph)
        print("PROCESSING NODE   :", n)
        print("-----------------------------------------------------------------------------------------")
        if self.gs(n)>=0:
            mc,dict=self.mcofn(n)
            print(mc,dict)
            self.sh(n,mc)
            self.ss(n,len(dict))
            print("h1",dict)
            solved=True
            for d in dict:
                self.parent[d]=n
                if self.gs(d)!=-1:
                    solved=solved & False
                if solved==True:
                    self.ss(n,-1)
                    print("here",d)
                    self.solutionGraph[n]=d
                if n!=self.start:
                    self.aoStar(self.parent[n],True)
                if bT==False:
                    for d in dict:
#                         self.ss(d,0)
                        self.aoStar(d,False)
h1 = {'A': 1, 'B': 6, 'C': 2, 'D': 12, 'E': 2, 'F': 1, 'G': 5, 'H': 7, 'I': 7, 'J': 1}
graph1 = {
    'A': [[('B', 1), ('C', 1)], [('D', 1)]],
    'B': [[('G', 1)], [('H', 1)]],
    'C': [[('J', 1)]],
    'D': [[('E', 1), ('F', 1)]],
    'G': [[('I', 1)]]   
}
G1= Graph(graph1, h1, 'A')
G1.applyAOStar()  
G1.printSolution()
```
